```
def agregar_empleado(lista_empleados, nombre):
    if isinstance(nombre, str) and 2 <= len(nombre) <= 50:
        lista_empleados.append(nombre.title())
        print(f"Empleado agregado: {nombre.title()}")
    else:
        print("Nombre de empleado inválido, debe tener entre 2 y 50 caracteres")

def mostrar_empleados(lista_empleados):
    for empleado in lista_empleados:
        print(f"-> {empleado}")

def modificar_empleado(lista_empleados, indice, nuevo_nombre):
    if not isinstance(nuevo_nombre, str) or not (2 <= len(nuevo_nombre) <= 50):
        print("Nombre de empleado inválido")
        return

    if 0 <= indice < len(lista_empleados):
        original = lista_empleados[indice]
        lista_empleados[indice] = nuevo_nombre.title()
        print(f"El empleado {original} fue reemplazado por {nuevo_nombre.title()}")
    else:
        print("Índice inválido")

def borrar_empleado(lista_empleados, indice):
    if 0 <= indice < len(lista_empleados):
        eliminado = lista_empleados.pop(indice)
        print(f"El empleado {eliminado} fue eliminado")
    else:
        print("Índice inválido")

def main():
    empleados = ["juan", "maria", "carlos", "andres"]

    print("Empleados actuales:")
    mostrar_empleados(empleados)

    agregar_empleado(empleados, "laura")
    mostrar_empleados(empleados)

    modificar_empleado(empleados, 2, "fernando")
    mostrar_empleados(empleados)

    borrar_empleado(empleados, 0)
    mostrar_empleados(empleados)

if __name__ == "__main__":
    main()
```
