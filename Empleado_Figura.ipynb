```
class Empleado:
    def __init__(self,nombre: str):
        if not Empleado.validar_nombre(nombre):
            raise ValueError("El nombre debe ser texto valido")
        self.nombre = nombre
    def trabajar(self) -> None:
        print(f"{self.nombre} esta realizando tareas generales")
    @staticmethod
    def validar_nombre(nombre: str) -> bool:
      return isinstance(nombre, str) and nombre.strip() !=""
class Gerente(Empleado):
    def trabajar(self) -> None:
      print(f"{self.nombre} esta cordinado el equipo y gestionado proyectos")
class Desarrollador(Empleado):
    def trabajar(self) -> None:
      print(f"{self.nombre} esta escribiendo codigo en phyton y resolviendo problemas de codigo")
def ejecutar_tarea(empleado: Empleado) -> None:
    empleado.trabajar()
def main() -> None:
    print("Es valido 'Eduardo' ? ->", Empleado.validar_nombre('Eduardo'))
    print("Es valido '' ? ->", Empleado.validar_nombre(''))
    gerente = Gerente("sofia")
    desarrollador = Desarrollador ("Jose")
    ejecutar_tarea(gerente)
    ejecutar_tarea(desarrollador)
if __name__ == "__main__":
  main()

```

```
from abc import ABC, abstractmethod
class Figura(ABC):
    @abstractmethod
    def calcular_area(self):
      pass
class Circulo(Figura):
    def __init__(self,radio):
        self.radio = radio
    def calcular_area(self):
        return 3.1416 * (self.radio ** 2)

class Cuadrado(Figura):
    def __init__(self,lado):
        self.lado = lado
    def calcular_area(self):
        return self.lado ** 2

def mostra_area(figura : Figura):
    print(f"Area: {figura.calcular_area():,.2f}")

def main():
  circulo = Circulo(5)
  mostra_area(circulo)
  cuadrado = Cuadrado(3)
  mostra_area(cuadrado)


if __name__ == "__main__":
  main()

```
