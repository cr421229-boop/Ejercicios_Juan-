```
class Operaciones():
    def __init__(self, numero1, numero2):
        self.numero1 = numero1
        self.numero2 = numero2
    def calcular(self):
      print("ALERTA vamos a iniciar operacion atentos")
class Suma(Operaciones):
    def __init__(self, numero1, numero2):
        super().__init__(numero1, numero2)
        print("preparando la operacion suma")

    def calcular(self):
        super().calcular()
        resultado = self.numero1 + self.numero2
        print(f"Mostrara el resultado de la suma es :{resultado}")
class Resta(Operaciones):
    def __init__(self, numero1, numero2):
        super().__init__(numero1, numero2)
        print("preparando resta")
    def calcular(self):
      super().calcular()
      resultado = self.numero1 - self.numero2
      print(f"mostrar resultado de la resta es: {resultado}")
class Multiplicacion(Operaciones):
    def __init__(self, numero1, numero2):
        super().__init__(numero1, numero2)
        print("preparando multiplicacion")
    def calcular(self):
      super().calcular()
      resultado = self.numero1 * self.numero2
      print(f"mostrar resultado de la multiplicacion es: {resultado}")
class Divicion(Operaciones):
    def __init__(self, numero1, numero2):
        super().__init__(numero1, numero2)
        print("preparando multiplicacion")
    def calcular(self):
      super().calcular()
      resultado = self.numero1 / self.numero2
      print(f"mostrar resultado de la divicion es: {resultado}")
class Potencia(Operaciones):
    def __init__(self, numero1, numero2):
        super().__init__(numero1, numero2)
        print("preparando potencia")
    def calcular(self):
      super().calcular()
      resultado = self.numero1 ** self.numero2
      print(f"mostrar resultado de la potencia es: {resultado}")


def main():
    operacion1 = Suma(20,30)
    operacion1.calcular()
    print("fin algoritmo de suma")
    operacion2 = Resta(40,50)
    operacion2.calcular()
    print("fin algoritmo resta")
    operacion3 = Multiplicacion(5,50)
    operacion3.calcular()
    print("fin algoritmo multiplicacion")
    operacion4 = Divicion(30,20)
    operacion4.calcular()
    print("Fin algoritmo divicion")
    operacion5 = Potencia(30,50)
    operacion5.calcular()
    print("fin algoritmo potencia")

if __name__ == "__main__":
  main()


```

```
class Operacion:
    def __init__(self,a, b):
        self.a = a
        self.b = b
    def validar(self):
        if not isinstance(self.a, (int,float)) or not isinstance(self.b, (int,float)):
            raise ValueError("Los valores deben ser numericos")
    def mostrar_datos(self):
        print(f"Los datos proporcionados son: {self.a} y {self.b}")
    def calcular(self):
        print("REALIZAR LA OPERACION MATEMATICA")
class Suma(Operacion):
    def calcular(self):
        super().validar()
        super().mostrar_datos()
        super().calcular()
        resultado = self.a + self.b
        print(f"el resultado de la suma es: {resultado}")
class Resta(Operacion):
      def calcular(self):
          super().validar()
          super().mostrar_datos()
          super().calcular()
          resultado = self.a - self.b
          print(f"el resultado de la resta es {resultado}")
class Multiplicacion(Operacion):
      def calcular(self):
          super().validar()
          super().mostrar_datos()
          super().calcular()
          resultado = self.a * self.b
          print(f"el rsultado de la multiplicacion es {resultado}")
class Division(Operacion):
      def calcular(self):
        super().validar()
        super().mostrar_datos()
        super().calcular()
        if self.b == 0:
          print("No se puede dividir entre cero")
        else:
          resultado = self.a / self.b
        print(f"el resultado de la divicion es {resultado}")
class Potencia(Operacion):
      def calcular(self):
        super().validar()
        super().mostrar_datos()
        super().calcular()
        if self.b >3:
          print("la potencia solo se puede hasta 3")
        else:
          resultado = self.a ** self.b
          print(f"el resultado de la potencia es: {resultado}")


def main():
    num1 = int(input("ingrese el primer numero"))
    num2 = int(input("ingrese el segundo numero"))
    print("OPERACION SUMA")
    operacion1 = Suma(num1,num2)
    operacion1.calcular()


    num3 = int(input("ingrese el primer numero"))
    num4 = int(input("ingrese el segundo numero"))
    print("OPERACION RESTA")
    operacion2 = Resta(num3,num4)
    operacion2.calcular()

    num5 = int(input("ingrese el primer numero"))
    num6 = int(input("ingrese el segundo numero"))
    print("OPERACION MULTIPLICACION")
    operacion3 = Multiplicacion(num5,num6)
    operacion3.calcular()


    num7 = int(input("ingrese el primer numero"))
    num8 = int(input("ingrese el segundo numero"))
    print("OPERACION DIVISION")
    operacion4 = Division(num7,num8)
    operacion4.calcular()


    num9 = int(input("ingrese el primer numero"))
    num10 = int(input("ingrese el segundo numero"))
    print("OPERACION POTENCIA")
    operacion5 = Potencia(num9,num10)
    operacion5.calcular()

if __name__ == "__main__":
    main()
```
