```
def agregar_mascota(lista_mascotas, nombre):
    if isinstance(nombre, str) and 2 <= len(nombre) <= 50:
        lista_mascotas.append(nombre.title())
        print(f"Mascota agregada: {nombre.title()}")
    else:
        print("Nombre de mascota inválido, debe tener entre 2 y 50 caracteres")

def mostrar_mascotas(lista_mascotas):
    for mascota in lista_mascotas:
        print(f"-> {mascota}")

def modificar_mascota(lista_mascotas, indice, nuevo_nombre):
    if not isinstance(nuevo_nombre, str) or not (2 <= len(nuevo_nombre) <= 50):
        print("Nombre de mascota inválido")
        return

    if 0 <= indice < len(lista_mascotas):
        original = lista_mascotas[indice]
        lista_mascotas[indice] = nuevo_nombre.title()
        print(f"La mascota {original} fue reemplazada por {nuevo_nombre.title()}")
    else:
        print("Índice inválido")

def borrar_mascota(lista_mascotas, indice):
    if 0 <= indice < len(lista_mascotas):
        eliminada = lista_mascotas.pop(indice)
        print(f"La mascota {eliminada} fue eliminada")
    else:
        print("Índice inválido")

def main():
    mascotas = ["firulais", "max", "luna", "rocky"]

    print("Mascotas actuales:")
    mostrar_mascotas(mascotas)

    agregar_mascota(mascotas, "nala")
    mostrar_mascotas(mascotas)

    modificar_mascota(mascotas, 2, "simba")
    mostrar_mascotas(mascotas)

    borrar_mascota(mascotas, 1)
    mostrar_mascotas(mascotas)

if __name__ == "__main__":
    main()
```
