```
from abc import ABC, abstractmethod

class Tienda(ABC):
    def __init__(self, nombre: str) -> None:
        self.nombre = nombre

    @property
    def nombre(self) -> str:
        return self._nombre

    @nombre.setter
    def nombre(self, valor: str) -> None:
        if isinstance(valor, str) and valor.strip():
            self._nombre = valor.strip()
        else:
            raise ValueError("El nombre debe ser válido")

    @abstractmethod
    def vender(self) -> None:
        pass


class TiendaRopa(Tienda):
    def vender(self) -> None:
        print(f"{self.nombre} está vendiendo ropa")


class TiendaElectronica(Tienda):
    def vender(self) -> None:
        print(f"{self.nombre} está vendiendo dispositivos electrónicos")


class TiendaAlimentos(Tienda):
    def vender(self) -> None:
        print(f"{self.nombre} está vendiendo alimentos")


class Libreria(Tienda):
    def vender(self) -> None:
        print(f"{self.nombre} está vendiendo libros")


class Farmacia(Tienda):
    def vender(self) -> None:
        print(f"{self.nombre} está vendiendo medicamentos")


def ejecutar_venta(tienda: Tienda) -> None:
    tienda.vender()


def main():
    print("***** SISTEMA DE TIENDAS *****\n")

    ropa = TiendaRopa("Moda Plus")
    electronica = TiendaElectronica("Tech Store")
    alimentos = TiendaAlimentos("Super Market")
    libreria = Libreria("Lectura Viva")
    farmacia = Farmacia("Salud Total")

    print("*** VENTAS ORIGINALES ***")
    ejecutar_venta(ropa)
    ejecutar_venta(electronica)
    ejecutar_venta(alimentos)
    ejecutar_venta(libreria)
    ejecutar_venta(farmacia)

    print("\n*** CAMBIANDO NOMBRES ***")
    ropa.nombre = "Fashion City"
    electronica.nombre = "Digital World"
    alimentos.nombre = "Fresh Market"
    libreria.nombre = "Libros y Más"
    farmacia.nombre = "Farmacenter"

    print("\n*** VENTAS ACTUALIZADAS ***")
    ejecutar_venta(ropa)
    ejecutar_venta(electronica)
    ejecutar_venta(alimentos)
    ejecutar_venta(libreria)
    ejecutar_venta(farmacia)


if __name__ == "__main__":
    main()
```
