```
from abc import ABC, abstractmethod

class Transporte(ABC):
    def __init__(self, nombre: str) -> None:
        self.nombre = nombre

    @property
    def nombre(self) -> str:
        return self._nombre

    @nombre.setter
    def nombre(self, valor: str) -> None:
        if isinstance(valor, str) and valor.strip():
            self._nombre = valor.strip()
        else:
            raise ValueError("El nombre debe ser válido")

    @abstractmethod
    def operar(self) -> None:
        pass


class Bus(Transporte):
    def operar(self) -> None:
        print(f"{self.nombre} está transportando pasajeros en la ciudad")


class Metro(Transporte):
    def operar(self) -> None:
        print(f"{self.nombre} está operando en líneas subterráneas")


class Taxi(Transporte):
    def operar(self) -> None:
        print(f"{self.nombre} está realizando carreras individuales")


class Tranvia(Transporte):
    def operar(self) -> None:
        print(f"{self.nombre} está circulando sobre rieles urbanos")


class Teleferico(Transporte):
    def operar(self) -> None:
        print(f"{self.nombre} está transportando personas por cable")


def ejecutar_transporte(transporte: Transporte) -> None:
    transporte.operar()


def main():
    print("***** SISTEMA DE TRANSPORTE *****\n")

    bus = Bus("Ruta 10")
    metro = Metro("Línea A")
    taxi = Taxi("Taxi 123")
    tranvia = Tranvia("Tranvía Centro")
    teleferico = Teleferico("Cable Norte")

    print("*** OPERACIONES ORIGINALES ***")
    ejecutar_transporte(bus)
    ejecutar_transporte(metro)
    ejecutar_transporte(taxi)
    ejecutar_transporte(tranvia)
    ejecutar_transporte(teleferico)

    print("\n*** CAMBIANDO NOMBRES ***")
    bus.nombre = "Ruta 20"
    metro.nombre = "Línea B"
    taxi.nombre = "Taxi 456"
    tranvia.nombre = "Tranvía Sur"
    teleferico.nombre = "Cable Sur"

    print("\n*** OPERACIONES ACTUALIZADAS ***")
    ejecutar_transporte(bus)
    ejecutar_transporte(metro)
    ejecutar_transporte(taxi)
    ejecutar_transporte(tranvia)
    ejecutar_transporte(teleferico)


if __name__ == "__main__":
    main()
```
