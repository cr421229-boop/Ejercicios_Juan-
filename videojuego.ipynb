```
def agregar_videojuego(lista_videojuegos, nombre):
    if isinstance(nombre, str) and 2 <= len(nombre) <= 50:
        lista_videojuegos.append(nombre.title())
        print(f"Videojuego agregado: {nombre.title()}")
    else:
        print("Nombre de videojuego inválido, debe tener entre 2 y 50 caracteres")

def mostrar_videojuegos(lista_videojuegos):
    for juego in lista_videojuegos:
        print(f"-> {juego}")

def modificar_videojuego(lista_videojuegos, indice, nuevo_nombre):
    if not isinstance(nuevo_nombre, str) or not (2 <= len(nuevo_nombre) <= 50):
        print("Nombre de videojuego inválido")
        return

    if 0 <= indice < len(lista_videojuegos):
        original = lista_videojuegos[indice]
        lista_videojuegos[indice] = nuevo_nombre.title()
        print(f"El videojuego {original} fue reemplazado por {nuevo_nombre.title()}")
    else:
        print("Índice inválido")

def borrar_videojuego(lista_videojuegos, indice):
    if 0 <= indice < len(lista_videojuegos):
        eliminado = lista_videojuegos.pop(indice)
        print(f"El videojuego {eliminado} fue eliminado")
    else:
        print("Índice inválido")

def main():
    videojuegos = ["minecraft", "fortnite", "fifa", "call of duty"]

    print("Videojuegos actuales:")
    mostrar_videojuegos(videojuegos)

    agregar_videojuego(videojuegos, "valorant")
    mostrar_videojuegos(videojuegos)

    modificar_videojuego(videojuegos, 2, "gta v")
    mostrar_videojuegos(videojuegos)

    borrar_videojuego(videojuegos, 1)
    mostrar_videojuegos(videojuegos)

if __name__ == "__main__":
    main()
```
