```
from abc import ABC, abstractmethod

class Profesion(ABC):
    def __init__(self, nombre: str) -> None:
        self.nombre = nombre

    @property
    def nombre(self) -> str:
        return self._nombre

    @nombre.setter
    def nombre(self, valor: str) -> None:
        if isinstance(valor, str) and valor.strip():
            self._nombre = valor.strip()
        else:
            raise ValueError("El nombre debe ser válido")

    @abstractmethod
    def trabajar(self) -> None:
        pass


class Medico(Profesion):
    def trabajar(self) -> None:
        print(f"{self.nombre} está atendiendo pacientes 🏥")


class Ingeniero(Profesion):
    def trabajar(self) -> None:
        print(f"{self.nombre} está diseñando proyectos ")


class Profesor(Profesion):
    def trabajar(self) -> None:
        print(f"{self.nombre} está enseñando en clase ")


class Policia(Profesion):
    def trabajar(self) -> None:
        print(f"{self.nombre} está protegiendo la ciudad ")


class Chef(Profesion):
    def trabajar(self) -> None:
        print(f"{self.nombre} está preparando comida ")


def ejecutar_trabajo(profesion: Profesion) -> None:
    profesion.trabajar()


def main():
    print("***** SISTEMA DE PROFESIONES *****\n")

    medico = Medico("Carlos")
    ingeniero = Ingeniero("Laura")
    profesor = Profesor("Andres")
    policia = Policia("Sofia")
    chef = Chef("Miguel")

    print("*** TRABAJOS ORIGINALES ***")
    ejecutar_trabajo(medico)
    ejecutar_trabajo(ingeniero)
    ejecutar_trabajo(profesor)
    ejecutar_trabajo(policia)
    ejecutar_trabajo(chef)

    print("\n*** CAMBIANDO NOMBRES ***")
    medico.nombre = "Julian"
    ingeniero.nombre = "Valentina"
    profesor.nombre = "Mateo"
    policia.nombre = "Daniela"
    chef.nombre = "Sebastian"

    print("\n*** TRABAJOS ACTUALIZADOS ***")
    ejecutar_trabajo(medico)
    ejecutar_trabajo(ingeniero)
    ejecutar_trabajo(profesor)
    ejecutar_trabajo(policia)
    ejecutar_trabajo(chef)


if __name__ == "__main__":
    main()
```
