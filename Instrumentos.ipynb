```
from abc import ABC, abstractmethod

class Instrumento(ABC):
    def __init__(self, nombre: str) -> None:
        self.nombre = nombre

    @property
    def nombre(self) -> str:
        return self._nombre

    @nombre.setter
    def nombre(self, valor: str) -> None:
        if isinstance(valor, str) and valor.strip():
            self._nombre = valor.strip()
        else:
            raise ValueError("El nombre debe ser válido")

    @abstractmethod
    def tocar(self) -> None:
        pass


class Guitarra(Instrumento):
    def tocar(self) -> None:
        print(f"{self.nombre} está tocando acordes ")


class Piano(Instrumento):
    def tocar(self) -> None:
        print(f"{self.nombre} está tocando melodías ")


class Bateria(Instrumento):
    def tocar(self) -> None:
        print(f"{self.nombre} está marcando el ritmo ")


class Violin(Instrumento):
    def tocar(self) -> None:
        print(f"{self.nombre} está interpretando música clásica ")


class Flauta(Instrumento):
    def tocar(self) -> None:
        print(f"{self.nombre} está produciendo sonidos suaves ")


def ejecutar_instrumento(instrumento: Instrumento) -> None:
    instrumento.tocar()


def main():
    print("***** SISTEMA DE INSTRUMENTOS *****\n")

    guitarra = Guitarra("Carlos")
    piano = Piano("Ana")
    bateria = Bateria("Luis")
    violin = Violin("Sofia")
    flauta = Flauta("Mateo")

    print("*** INTERPRETACIONES ORIGINALES ***")
    ejecutar_instrumento(guitarra)
    ejecutar_instrumento(piano)
    ejecutar_instrumento(bateria)
    ejecutar_instrumento(violin)
    ejecutar_instrumento(flauta)

    print("\n*** CAMBIANDO NOMBRES ***")
    guitarra.nombre = "Diego"
    piano.nombre = "Valeria"
    bateria.nombre = "Andres"
    violin.nombre = "Camila"
    flauta.nombre = "Julian"

    print("\n*** INTERPRETACIONES ACTUALIZADAS ***")
    ejecutar_instrumento(guitarra)
    ejecutar_instrumento(piano)
    ejecutar_instrumento(bateria)
    ejecutar_instrumento(violin)
    ejecutar_instrumento(flauta)


if __name__ == "__main__":
    main()
```
