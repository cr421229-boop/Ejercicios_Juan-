```
def agregar_ciudad(lista_ciudades, nombre):
    if isinstance(nombre, str) and 2 <= len(nombre) <= 50:
        lista_ciudades.append(nombre.title())
        print(f"Ciudad agregada: {nombre.title()}")
    else:
        print("Nombre de ciudad inválido, debe tener entre 2 y 50 caracteres")

def mostrar_ciudades(lista_ciudades):
    for ciudad in lista_ciudades:
        print(f"-> {ciudad}")

def modificar_ciudad(lista_ciudades, indice, nuevo_nombre):
    if not isinstance(nuevo_nombre, str) or not (2 <= len(nuevo_nombre) <= 50):
        print("Nombre de ciudad inválido")
        return

    if 0 <= indice < len(lista_ciudades):
        original = lista_ciudades[indice]
        lista_ciudades[indice] = nuevo_nombre.title()
        print(f"La ciudad {original} fue reemplazada por {nuevo_nombre.title()}")
    else:
        print("Índice inválido")

def borrar_ciudad(lista_ciudades, indice):
    if 0 <= indice < len(lista_ciudades):
        eliminada = lista_ciudades.pop(indice)
        print(f"La ciudad {eliminada} fue eliminada")
    else:
        print("Índice inválido")

def main():
    ciudades = ["bogota", "medellin", "cali", "cartagena"]

    print("Ciudades actuales:")
    mostrar_ciudades(ciudades)

    agregar_ciudad(ciudades, "barranquilla")
    mostrar_ciudades(ciudades)

    modificar_ciudad(ciudades, 2, "pereira")
    mostrar_ciudades(ciudades)

    borrar_ciudad(ciudades, 1)
    mostrar_ciudades(ciudades)

if __name__ == "__main__":
    main()
```
