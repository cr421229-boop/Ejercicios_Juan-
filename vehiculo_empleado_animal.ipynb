```
class Empleado:
  def __init__(self, nombre, salario):
      self.nombre = nombre
      self.salario = salario
  def trabajar(self):
    print(f"{self.nombre} esta realizando un trabajo en la empresa")
class Gerente(Empleado):
  def __init__(self, nombre, salario, departamento):
      super().__init__(nombre, salario)
      self.departamento = departamento
  def mostrar_info(self):
    print(f"nombre: atributo del padre:{self.nombre}")
    print(f"Salario: Atributo del pdre: {self.salario}")
    print(f"departamento: Atributo del hijo:{self.departamento}")

class Vendedor(Empleado):
  def __init__(self, nombre, salario,ciudad, comision):
    super().__init__(nombre, salario)
    self.ciudad = ciudad
    self.comision = comision


  def mostrar_info(self):

    print(f"nombre: Atributo del hijo:{self.nombre}")
    print(f"salario: Atributo del hijo:{self.salario}")
    print(f"cuidad: Atributo del hijo:{self.ciudad}")
    print(f"comision: Atributo del hijo:{self.comision}")

def main():
  Gerente1 = Gerente("carlos",123333,"software")
  Gerente1.mostrar_info()

  Vendedor1 = Vendedor("Juan", 1111, "lllll", 22222)
  Vendedor1.mostrar_info()
if __name__ == "__main__":
  main()
```
