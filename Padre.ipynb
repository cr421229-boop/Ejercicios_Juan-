```python
class Padre:
 def __init__(self,mensaje):
    self.mensaje = mensaje

class Hijo(Padre):
 def __init__(self,mensaje,nombre):
  super().__init__(mensaje)
  self.nombre = nombre

def main():
  obj = Hijo("Hola desde mensaje padre ","Pedro")
  print(f"{obj.mensaje},{obj.nombre}\n")
  print(obj.mensaje)
  print(obj.nombre)
if __name__ == "__main__":
  main()
```

```python
class Empleado:
  def __init__(self,nombre, salario):
    self.nombre = nombre
    self.salario = salario
class Gerente(Empleado):
  def __init__(self, nombre,salario,departamento):
    super().__init__(nombre, salario)
    self.departamento = departamento
  def mostrar_info(self):
   print(f"Nombre:{self.nombre}, salario: ${self.salario:,.0f}")
   print(f"Departamento: {self.departamento}")
def main():
  gerente1 = Gerente("Pedro Moreno",243334,"Inovacion")
  gerente1.mostrar_info()

if __name__ == "__main__":
  main()




```
