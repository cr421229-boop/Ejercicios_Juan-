```python
def agregar_muebles(lista_muebles, mueble):
    if isinstance(mueble, str) and 2 <= len(mueble) <= 50:
        lista_muebles.append(mueble.title())
        print(f"Mueble agregado: {mueble.title()}")
    else:
        print("El mueble debe tener entre 2 y 50 caracteres")


def mostrar_muebles(lista_muebles):
    if not lista_muebles:
        print("No hay muebles registrados")
    else:
        for i, muebles in enumerate(lista_muebles):
            print(f"{i}. {muebles}")


def modificar_muebles(lista_muebles, indice, nuevo_mueble):
    if isinstance(nuevo_mueble, str) and 2 <= len(nuevo_mueble) <= 50:
        if 0 <= indice < len(lista_muebles):
            lista_muebles[indice] = nuevo_mueble.title()
            print("Mueble modificado correctamente")
        else:
            print("Índice fuera de rango")
    else:
        print("El nuevo mueble debe tener entre 2 y 50 caracteres")


def eliminar_muebles(lista_muebles, indice):
    if 0 <= indice < len(lista_muebles):
        eliminado = lista_muebles.pop(indice)
        print(f"Mueble eliminado: {eliminado}")
    else:
        print("Índice fuera de rango")


def main():
    lista_muebles = []

    while True:
        print("\n--- MENÚ ---")
        print("1. Agregar mueble")
        print("2. Mostrar muebles")
        print("3. Modificar mueble")
        print("4. Eliminar mueble")
        print("5. Salir")

        opcion = input("Seleccione una opción: ")

        if opcion == "1":
            mueble = input("Ingrese el nombre del mueble: ")
            agregar_muebles(lista_muebles, mueble)

        elif opcion == "2":
            mostrar_muebles(lista_muebles)

        elif opcion == "3":
            try:
                indice = int(input("Ingrese el índice del mueble a modificar: "))
                nuevo = input("Ingrese el nuevo nombre: ")
                modificar_muebles(lista_muebles, indice, nuevo)
            except ValueError:
                print("Debe ingresar un número válido")

        elif opcion == "4":
            try:
                indice = int(input("Ingrese el índice del mueble a eliminar: "))
                eliminar_muebles(lista_muebles, indice)
            except ValueError:
                print("Debe ingresar un número válido")

        elif opcion == "5":
            print("Saliendo del sistema...")
            break

        else:
            print("Opción inválida")


# Ejecutar programa
if __name__ == "__main__":
    main()
```
