```python
class Libro:
    def __init__(self, titulo, precio):
      self.__titulo = titulo
      self.__precio = precio

    @property
    def precio(self):
      return self.__precio
    @property
    def titulo(self):
     return self.__titulo
    @precio.setter
    def precio(self, nuevo_precio):
      if isinstance(nuevo_precio,(int, float)) and nuevo_precio > 0:
        self.__precio = nuevo_precio
      else:
        print("Debe ingresar un número valido en el precio")

    @titulo.setter
    def titulo(self,nuevo_titulo):
        if isinstance(nuevo_titulo, str):
          self.__titulo = nuevo_titulo
        else:
          print("debe ingresar un texto valido en el titulo")
    def mostrar_informacion(self):
      print(f"\nel titulo del libro es:{self.__titulo}")
      print(f"el precio del libro es:{self.__precio}\n")

def main():
  print("=====CREACION DE OBJETOS EM POO====")

  libro1 = Libro("el principito",67000)
  libro2 = Libro("el quijote de la mancha",25000 )
  print("El nombre del libro es:")
  print( libro1.titulo)
  print("el precio del libro es: ", libro1.precio)
  print("el nombre del libro es:  ", libro1.titulo, "precio:", libro1.precio )
  print("todos los datos")
  libro1.mostrar_informacion()

  print("actualizaremos el titulo\n")
  libro1.titulo = input("\ningrese el nuevo titulo ==")
  print("Titulo actualizado ahora el libro 1 se llama:", libro1.titulo)
  print("el precio del libro es:", libro1.precio,"actualizaremos este precio por otro\n")
  libro1.precio = input("ingrese el nuevo precio del libro 1:")
  print("\n precio actualizado ahora el libro 1 cuesta:", libro1.precio)
  print("\nLos nuevos datos del libro 1 son")
  libro1.mostrar_informacion()









if __name__ =="__main__":
   main()
```
