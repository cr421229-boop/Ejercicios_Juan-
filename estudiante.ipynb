```
def agregar_estudiante(lista_estudiantes, nombre):
    if isinstance(nombre, str) and 2 <= len(nombre) <= 50:
        lista_estudiantes.append(nombre.title())
        print(f"Estudiante agregado: {nombre.title()}")
    else:
        print("Nombre de estudiante inválido, debe tener entre 2 y 50 caracteres")

def mostrar_estudiantes(lista_estudiantes):
    for estudiante in lista_estudiantes:
        print(f"-> {estudiante}")

def modificar_estudiante(lista_estudiantes, indice, nuevo_nombre):
    if not isinstance(nuevo_nombre, str) or not (2 <= len(nuevo_nombre) <= 50):
        print("Nombre de estudiante inválido")
        return

    if 0 <= indice < len(lista_estudiantes):
        original = lista_estudiantes[indice]
        lista_estudiantes[indice] = nuevo_nombre.title()
        print(f"El estudiante {original} fue reemplazado por {nuevo_nombre.title()}")
    else:
        print("Índice inválido")

def borrar_estudiante(lista_estudiantes, indice):
    if 0 <= indice < len(lista_estudiantes):
        eliminado = lista_estudiantes.pop(indice)
        print(f"El estudiante {eliminado} fue eliminado")
    else:
        print("Índice inválido")

def main():
    estudiantes = ["ana", "carlos", "luisa", "mateo"]

    print("Estudiantes actuales:")
    mostrar_estudiantes(estudiantes)

    agregar_estudiante(estudiantes, "sofia")
    mostrar_estudiantes(estudiantes)

    modificar_estudiante(estudiantes, 0, "valentina")
    mostrar_estudiantes(estudiantes)

    borrar_estudiante(estudiantes, 2)
    mostrar_estudiantes(estudiantes)

if __name__ == "__main__":
    main()
```
