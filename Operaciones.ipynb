```
class Operaciones():
    def __init__(self, numero1, numero2):
        self.numero1 = numero1
        self.numero2 = numero2
    def calcular(self):
      print("ALERTA vamos a iniciar operacion atentos")
class Suma(Operaciones):
    def __init__(self, numero1, numero2):
        super().__init__(numero1, numero2)
        print("preparando la operacion suma")

    def calcular(self):
        super().calcular()
        resultado = self.numero1 + self.numero2
        print(f"Mostrara el resultado de la suma es :{resultado}")
class Resta(Operaciones):
    def __init__(self, numero1, numero2):
        super().__init__(numero1, numero2)
        print("preparando resta")
    def calcular(self):
      super().calcular()
      resultado = self.numero1 - self.numero2
      print(f"mostrar resultado de la resta es: {resultado}")
class Multiplicacion(Operaciones):
    def __init__(self, numero1, numero2):
        super().__init__(numero1, numero2)
        print("preparando multiplicacion")
    def calcular(self):
      super().calcular()
      resultado = self.numero1 * self.numero2
      print(f"mostrar resultado de la multiplicacion es: {resultado}")
class Divicion(Operaciones):
    def __init__(self, numero1, numero2):
        super().__init__(numero1, numero2)
        print("preparando multiplicacion")
    def calcular(self):
      super().calcular()
      resultado = self.numero1 / self.numero2
      print(f"mostrar resultado de la divicion es: {resultado}")
class Potencia(Operaciones):
    def __init__(self, numero1, numero2):
        super().__init__(numero1, numero2)
        print("preparando potencia")
    def calcular(self):
      super().calcular()
      resultado = self.numero1 ** self.numero2
      print(f"mostrar resultado de la potencia es: {resultado}")


def main():
    operacion1 = Suma(20,30)
    operacion1.calcular()
    print("fin algoritmo de suma")
    operacion2 = Resta(30,50)
    operacion2.calcular()
    print("fin algoritmo resta")
    operacion3 = Multiplicacion(5,50)
    operacion3.calcular()
    print("fin algoritmo multiplicacion")
    operacion4 = Divicion(30,20)
    operacion4.calcular()
    print("Fin algoritmo divicion")
    operacion5 = Potencia(30,50)
    operacion5.calcular()
    print("fin algoritmo potencia")

if __name__ == "__main__":
  main()


```
