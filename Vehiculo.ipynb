```
from abc import ABC, abstractmethod

class Vehiculo(ABC):
    def __init__(self, nombre: str) -> None:
        self.nombre = nombre

    @property
    def nombre(self) -> str:
        return self._nombre

    @nombre.setter
    def nombre(self, valor: str) -> None:
        if isinstance(valor, str) and valor.strip():
            self._nombre = valor.strip()
        else:
            raise ValueError("El nombre debe ser válido")

    @abstractmethod
    def moverse(self) -> None:
        pass


class Carro(Vehiculo):
    def moverse(self) -> None:
        print(f"{self.nombre} está conduciendo por la carretera 🚗")


class Moto(Vehiculo):
    def moverse(self) -> None:
        print(f"{self.nombre} está acelerando rápidamente 🏍️")


class Avion(Vehiculo):
    def moverse(self) -> None:
        print(f"{self.nombre} está volando por el cielo ✈️")


class Barco(Vehiculo):
    def moverse(self) -> None:
        print(f"{self.nombre} está navegando en el mar 🚢")


class Bicicleta(Vehiculo):
    def moverse(self) -> None:
        print(f"{self.nombre} está pedaleando 🚴")


def ejecutar_movimiento(vehiculo: Vehiculo) -> None:
    vehiculo.moverse()


def main():
    print("***** SISTEMA DE VEHÍCULOS *****\n")

    carro = Carro("Toyota")
    moto = Moto("Yamaha")
    avion = Avion("Boeing")
    barco = Barco("Titanic")
    bicicleta = Bicicleta("GW")

    print("*** MOVIMIENTOS ORIGINALES ***")
    ejecutar_movimiento(carro)
    ejecutar_movimiento(moto)
    ejecutar_movimiento(avion)
    ejecutar_movimiento(barco)
    ejecutar_movimiento(bicicleta)

    print("\n*** CAMBIANDO NOMBRES ***")
    carro.nombre = "Mazda"
    moto.nombre = "Suzuki"
    avion.nombre = "Airbus"
    barco.nombre = "Poseidón"
    bicicleta.nombre = "Trek"

    print("\n*** MOVIMIENTOS ACTUALIZADOS ***")
    ejecutar_movimiento(carro)
    ejecutar_movimiento(moto)
    ejecutar_movimiento(avion)
    ejecutar_movimiento(barco)
    ejecutar_movimiento(bicicleta)


if __name__ == "__main__":
    main()
```
