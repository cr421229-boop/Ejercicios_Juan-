```
from abc import ABC, abstractmethod

class Empleado(ABC):
    def __init__(self, nombre: str) -> None:
        self.nombre = nombre  # usa el setter

    @property
    def nombre(self) -> str:
        return self._nombre

    @nombre.setter
    def nombre(self, valor: str) -> None:
        if isinstance(valor, str) and valor.strip():
            self._nombre = valor.strip()
        else:
            raise ValueError("El nombre debe ser un texto válido")

    @abstractmethod
    def trabajar(self) -> None:
        pass


# CLASES QUE YA TENÍA
class Gerente(Empleado):
    def trabajar(self) -> None:
        print(f"{self.nombre} está realizando la administración de la empresa")


class Desarrollador(Empleado):
    def trabajar(self) -> None:
        print(f"{self.nombre} está programando una solución en Python")

#NUEVAS CLASES
class Vendedor(Empleado):
    def trabajar(self) -> None:
        print(f"{self.nombre} está vendiendo productos")


class Tecnico(Empleado):
    def trabajar(self) -> None:
        print(f"{self.nombre} está reparando equipos")


class Secretario(Empleado):
    def trabajar(self) -> None:
        print(f"{self.nombre} está organizando documentos")


# POLIMORFISMO
def ejecutar_tarea(empleado: Empleado) -> None:
    empleado.trabajar()


def main():
    print("****** INICIO DEL PROGRAMA DE EMPLEADOS EMPRESA SENA *****\n")

    # CREAR EMPLEADOS
    desarrollador1 = Desarrollador("Alexander Joaquin")
    gerente1 = Gerente("Pedrito Perez")
    vendedor1 = Vendedor("Maria Gomez")
    tecnico1 = Tecnico("Carlos Ruiz")
    secretario1 = Secretario("Ana Torres")

    print("*** DATOS ORIGINALES ***")
    ejecutar_tarea(desarrollador1)
    ejecutar_tarea(gerente1)
    ejecutar_tarea(vendedor1)
    ejecutar_tarea(tecnico1)
    ejecutar_tarea(secretario1)

    # CAMBIAR NOMBRES
    print("\n*** CAMBIANDO NOMBRES ***")
    desarrollador1.nombre = "Vanessa Etoco"
    gerente1.nombre = "Luis Martinez"
    vendedor1.nombre = "Laura Diaz"
    tecnico1.nombre = "Andres Castro"
    secretario1.nombre = "Camila Rojas"

    print("\n*** DATOS ACTUALIZADOS ***")
    ejecutar_tarea(desarrollador1)
    ejecutar_tarea(gerente1)
    ejecutar_tarea(vendedor1)
    ejecutar_tarea(tecnico1)
    ejecutar_tarea(secretario1)


if __name__ == "__main__":
    main()
```

```
from abc import ABC, abstractmethod

# INTERFAZ (Movible)
class Movible(ABC):
    @abstractmethod
    def mover(self) -> None:
        pass



# CLASE ABSTRACTA VENDEDOR
class Vendedor(ABC):
    def __init__(self, nombre: str, apellido: str) -> None:
        self.__nombre = ""
        self.__apellido = ""
        self.nombre = nombre
        self.apellido = apellido

    @property
    def nombre(self) -> str:
        return self.__nombre

    @nombre.setter
    def nombre(self, valor: str) -> None:
        if isinstance(valor, str) and valor.strip():
            self.__nombre = valor.strip().title()
        else:
            raise ValueError("Nombre inválido")

    @property
    def apellido(self) -> str:
        return self.__apellido

    @apellido.setter
    def apellido(self, valor: str) -> None:
        if isinstance(valor, str) and valor.strip():
            self.__apellido = valor.strip().title()
        else:
            raise ValueError("Apellido inválido")

    @abstractmethod
    def vender(self) -> None:
        pass


# VENDEDORES ESPECÍFICOS
class VendedorTienda(Vendedor):
    def vender(self) -> None:
        print(f"{self.nombre} {self.apellido} vende en una tienda física")


class VendedorOnline(Vendedor):
    def vender(self) -> None:
        print(f"{self.nombre} {self.apellido} vende por internet")


class VendedorPorMayor(Vendedor):
    def vender(self) -> None:
        print(f"{self.nombre} {self.apellido} vende al por mayor")


# SUBCLASES
class VendedorDomicilio(Vendedor, Movible):
    def vender(self) -> None:
        print(f"{self.nombre} {self.apellido} vende a domicilio")

    def mover(self) -> None:
        print(f"{self.nombre} se desplaza por la ciudad entregando pedidos")


class VendedorTelemercaderista(Vendedor):
    def vender(self) -> None:
        print(f"{self.nombre} {self.apellido} vende por llamadas telefónicas")


# DOS MÁS
class VendedorFreelance(Vendedor, Movible):
    def vender(self) -> None:
        print(f"{self.nombre} {self.apellido} vende de forma independiente")

    def mover(self) -> None:
        print(f"{self.nombre} se mueve entre clientes buscando ventas")


class VendedorExportador(Vendedor):
    def vender(self) -> None:
        print(f"{self.nombre} {self.apellido} vende productos al extranjero")


# FUNCIONES POLIMÓRFICAS
def vender(vendedor: Vendedor) -> None:
    vendedor.vender()


def mover(vendedor: Movible) -> None:
    vendedor.mover()


# MAIN
def main():
    try:
        vendedores = [
            VendedorTienda("Ana", "Lopez"),
            VendedorOnline("Carlos", "Perez"),
            VendedorPorMayor("Luis", "Gomez"),
            VendedorDomicilio("Maria", "Torres"),
            VendedorTelemercaderista("Sofia", "Ramirez"),
            VendedorFreelance("Pedro", "Diaz"),
            VendedorExportador("Laura", "Castro")
        ]

        print("=== SISTEMA DE VENDEDORES ===\n")

        print("=== VENTAS ===")
        for v in vendedores:
            vender(v)

        print("\n=== MOVIMIENTO ===")
        for v in vendedores:
            if isinstance(v, Movible):
                mover(v)

        print("\n=== ACTUALIZACIÓN DE DATOS ===")
        vendedores[0].nombre = "Camila"
        print(f"Nuevo nombre: {vendedores[0].nombre} {vendedores[0].apellido}")

        print("\n=== DATOS ACTUALIZADOS ===")
        for v in vendedores:
            vender(v)

    except ValueError as e:
        print(f"Error: {e}")


if __name__ == "__main__":
    main()
```
